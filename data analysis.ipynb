```python
# State based categorizing
states, counts = np.unique(df["state"],return_counts=True)
cat = {}
state = [0]*len(states)
df2 = pd.DataFrame()
min_magnitude = df.groupby(["state"])["magnitudo"].min()
max_magnitude = df.groupby(["state"])["magnitudo"].max()
mean_magnitude = df.groupby(["state"])["magnitudo"].mean()
tsunamis = df.groupby(["state"])["tsunami"].sum()
count = df.groupby(["state"]).size()
df2["state"] = pd.Series(states)
df2["min_magnitude"] = df2["state"].apply(lambda state:min_magnitude[state])
df2["max_magnitude"] = df2["state"].apply(lambda state:max_magnitude[state])
df2["average_magnitude"] = df2["state"].apply(lambda state:mean_magnitude[state])
df2["tsunamis"] = df2["state"].apply(lambda state:tsunamis[state])
df2["count"] = df2["state"].apply(lambda state:count[state])
```

```python
df2
```

```python
# based on severity
severe = df[df["magnitudo"] >= 5.5] #Usually earthquakes with magnitude 5.5 or more have serious effects
less_severe = df[df["magnitudo"] < 5.5]  #Usually earthquakes with magnitude less than 5.5 are less sever
s_counts = pd.value_counts(severe["state"])
ls_counts = pd.value_counts(less_severe["state"])
df2["5.5_or_greater_count"] = df2["state"].apply(lambda state: s_counts[state] if state in s_counts else 0)
df2["less_severe_count"] = df2["state"].apply(lambda state: ls_counts[state] if state in ls_counts else 0)
```
